Import libraries and modules needed for stimulation signal

```python
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../Scripts")

import stim_util as su
from simulation import Simulation
```

Let's create a simulaton environment (this is not necessary though, but with this we do not have to import numpy)

```python
mySim = Simulation(period=10000, fs=100)
```

Create a stimulation signal

```python
stim_chirp, freq = su.gen_chirp(init_freq=.00001, init_time=0, end_freq=.01, end_time=mySim.t[-1], time_points=mySim.t)
stim_pulse = su.gen_pulse(init_time=500, on_width=1000, off_width=2000, time_points=mySim.t)
stim_sin = su.gen_sin(init_time=0, freq=.001, phase=0, time_points=mySim.t)
stim_slope = su.gen_slope(init_time=0, end_time=5000, time_points=mySim.t)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t, stim_chirp, 'k', linewidth=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t, stim_pulse, 'k', linewidth=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t, stim_sin, 'k', linewidth=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t, stim_slope, 'k', linewidth=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t, 2 * stim_sin + stim_chirp, 'k', linewidth=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(mySim.t,  2 * stim_slope * (stim_sin + stim_chirp), 'k', linewidth=3)
plt.show()
```
