In This notebook we discuss the concept of Temporal Interference

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import functions as fn
```

Define simulation parameters

```python
Sim_period = .2  # seconds
Sim_fs = 100000  # samples per second

Sim_dt = 1/Sim_fs
Sim_t = np.arange(0, Sim_period, Sim_dt)
```

Define the signal

```python
sig1 = fn.gen_sin(init_time=0, freq=100, phase=0, time_points=Sim_t)
sig2 = fn.gen_sin(init_time=0, freq=110, phase=0, time_points=Sim_t)
```

Define the style of the plot

```python
fig, ax = plt.subplots(figsize=(6, 2), dpi=500)

# remove the 
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# remove the x and y ticks
ax.set_yticks([],[])
ax.set_xticks([],[])

ax.plot(
    Sim_t, sig1, 'b',
#     Sim_t, sig2, 'k' ,
    lineWidth=2
)

save = True
# save the figure

if save:
    name = 'sig_1_f-100'
    folder = 'D:/TUM/Semester 3 - Holidays/Symposium/Presentation/Figures/TI/'
    path = folder + name
    plt.savefig(path + '.png', bbox_inches = 'tight', transparent=True)
```

### plot them together

```python
fig, ax = plt.subplots(figsize=(4, 2), dpi=500)

# remove the 
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# remove the x and y ticks
ax.set_yticks([],[])
ax.set_xticks([],[])

ax.plot(
#     Sim_t, sig1, 'b'
#     , Sim_t, sig2, 'k', 
    Sim_t, sig1 + sig2, 'r'
    , lineWidth=3)

save = True
# save the figure

if save:
    name = 'sig_AM_paper_alone_x2_f-100_thick'
    folder = 'D:/TUM/Semester 3 - Holidays/Symposium/Presentation/Figures/TI/'
    path = folder + name
    plt.savefig(path + '.png', bbox_inches = 'tight', transparent=True)
```

### Create the envelope using a sinusoid

```python
 from scipy.signal import hilbert, chirp
```

```python
signal = sig1 + sig2
signal_envelope = np.abs(hilbert(signal))
```

```python
fig, ax = plt.subplots(figsize=(4, 2), dpi=500)

# remove the 
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# remove the x and y ticks
ax.set_yticks([],[])
ax.set_xticks([],[])

ax.plot(
    Sim_t, signal_envelope, 'r',
    Sim_t, signal, 'r',
    lineWidth=1)

save = False
# save the figure

if save:
    name = 'sig_AM_paper_envelope_alone'
    folder = 'D:/TUM/Semester 3 - Holidays/Symposium/Presentation/Figures/TI/'
    path = folder + name
    plt.savefig(path + '.png', bbox_inches = 'tight', transparent=True)
```
