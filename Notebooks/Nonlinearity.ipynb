In this notebook we explore the effect of non-linearity on the electrical model

Import useful libraries

```python
import functions as fn
import numpy as np
import matplotlib.pyplot as plt
```

Define simulation parameters

```python
Sim_period = 100000  # ms
Sim_fs   = 10  # samples per ms

Sim_dt = 1/Sim_fs
Sim_t = np.arange(0,Sim_period,Sim_dt)
```

Define membrane properties

```python
g_leak = .1
g_Na = .001
g_L = 100
L = 2.5  # kH  -> since the simulation is in ms, then we need the time constants to be in 
C = 100  # mF
V_leak = 0
V_Na = 140

tau_L = g_L * L # * 1e3, because we want it in milisecond
tau_C = g_leak/(C * 1e-3)

print('Inductor time constant:', tau_L, 'ms')
print('Capacitor time constant:', tau_C, 's')
print('Resonance frequency:', 1000 * 1/(2*np.pi*np.sqrt(L*C)))

I_stim = 1

# non-linearity
k = 1
V_half = 0#2
V_min = -5 * I_stim
V_max = 5 * I_stim
Vm_range = np.ceil(V_max - V_min)

Vm = np.arange(V_min, V_max, .0001)
Boltzman_nl = k / (1 + np.exp(20/Vm_range*(V_half - Vm)))
plt.plot(Vm, Boltzman_nl**3)
plt.show()
```

### Impedance of the model vs. frequency

```python
f = np.arange(.00001,10,.00001)  # kHz
w = 2 * np.pi * f

R_L = 1 / g_L
R_leak = 1 / g_leak

z = (g_leak + 1j*w*C + 1/(1j*w*L + R_L)) ** -1
z_amp = np.sqrt(z.real**2 + z.imag**2)

z_L = 1j*w*L + R_L
z_L_amp = np.sqrt(z_L.real**2 + z_L.imag**2)

z_C = (1j * w * C + g_leak) ** -1
z_C_amp = np.sqrt(z_C.real**2 + z_C.imag**2)

plt.figure(figsize=(20,5))
plt.loglog(f * 1000, z_amp)
plt.loglog(f * 1000, z_L_amp)
plt.loglog(f * 1000, z_C_amp)
plt.grid()
plt.show()
```

### Stimulation signal

```python
I = I_stim * np.ones(Sim_t.shape)  # pA

I1 = fn.gen_sin(init_time=0,
                freq=1,
                phase=0,
                time_points=Sim_t)
I2 = fn.gen_sin(init_time=0,
                freq=1.01,
                phase=0,
                time_points=Sim_t)

I = I_stim * I1

# I1 = fn.gen_pulse(init_time=1000,
#                  on_width=100, # width of "ON" state
#                  off_width=100, # width of "OFF" state
#                  time_points=Sim_t)

# I2 = fn.gen_pulse(init_time=1000,
#                  on_width=1000, # width of "ON" state
#                  off_width=1000, # width of "OFF" state
#                  time_points=Sim_t)

# I = I_stim * I1 * I2

# Chirp stimulation
out, freq = fn.gen_chirp(Chirp_init_freq=.00000001, 
                         Chirp_init_time=0, 
                         Chirp_end_freq=.02, 
                         Chirp_end_time=Sim_t[-1], 
                         time_points=Sim_t)

I      = I_stim * out  # pA
```

### Parapeters to keep track of

```python
Vm = np.zeros(Sim_t.shape)
I_L = np.zeros(Sim_t.shape)
Boltzman_nl = np.zeros(Vm.shape)
```

### Run simulation

```python
for i in range(I.shape[0]-1):
    
    # Boltzmann non-linearity
    Boltzman_nl[i] = k / (1 + np.exp(20/Vm_range*(V_half - Vm[i])))
    
    # current through the leak branch
    I_leak = g_leak * (Vm[i] - V_leak)
    
    # current through the membrane capacitance
    I_Na = g_Na * (Boltzman_nl[i]**3) * (Vm[i] - V_Na)
    
    # current through the inductor
    I_L_ss = Vm[i] * g_L
    dILdt = (I_L_ss - I_L[i]) / tau_L
    I_L[i+1] = I_L[i] + dILdt * Sim_dt
    
    
#     I_L[i+1] = 0
#     I_leak = 0
#     I_Na = 0
    
    dVmdt = (I[i] - I_leak - I_Na - I_L[i]) / C
    Vm[i+1] = Vm[i] + dVmdt * Sim_dt
```

### Plot results

```python
plt.figure(figsize=(20,2))
plt.plot(Sim_t, I)
plt.show()

plt.figure(figsize=(20,5))
plt.plot(Sim_t, Vm)
plt.show()

plt.figure(figsize=(20,2))
plt.plot(Sim_t, I_L)
plt.show()

plt.figure(figsize=(20,2))
plt.plot(Sim_t, Boltzman_nl)
plt.show()
```
