Import useful libraries

```python
import functions as fn
import numpy as np
import matplotlib.pyplot as plt
```

Defining simulation parameters

```python
Sim_period = 100000
Sim_fs   = 10

Sim_dt = 1/Sim_fs
Sim_t = np.arange(0,Sim_period,Sim_dt)
```

Membrane parameters (constant over time)

```python
Vrest = 0
Rm = 1
Lm = 1000
Cm = 300

tau_C = Rm * Cm
tau_L = Lm / Rm
```

Membrane parameters (keep track over time)

```python
V_C = np.zeros(Sim_t.shape)
V_L = np.zeros(Sim_t.shape)
I_L = np.zeros(Sim_t.shape)
```

Stimulation Parameters

```python
I_stim = np.ones(Sim_t.shape)
```

Run the simulation

```python
for i in range(I_stim.shape[0]-1):
    
    # current through the capacitor
    dVCdt = (I_stim[i] * Rm - V_C[i]) / tau_C
    V_C[i+1] = V_C[i] + dVCdt * Sim_dt
       
    # current through the inductor
    dILdt = (I_stim[i] - I_L[i]) / tau_L
    I_L[i+1] = I_L[i] + dILdt * Sim_dt
    V_L[0] = I_stim[0] * Rm - I_L[0] * Rm  # added just t compensate for the skip of first element
    V_L[i+1] = I_stim[i+1] * Rm - I_L[i+1] * Rm
```

```python
plt.plot(Sim_t, V_C)
plt.show()
plt.plot(Sim_t, V_L)
plt.plot(Sim_t, I_L)
plt.show()
# plt.plot(t, V_C + V_L)
# plt.show()
```

```python
print(
    V_C[int(tau_C * Sim_fs)],
    V_L[int(tau_L * Sim_fs)]
)
```

```python
V_half = 0
k = 1
Boltzman_nl = k / (1 + np.exp(V_half - Vm))
```

## Demodulator using diode non-linearity

```python
g_leak = 1
g_Na = 1
C = 100
```

```python
Vm = np.zeros(Sim_t.shape)
diode_nl = np.zeros(Sim_t.shape)

V_half = 0
diode_nl = np.zeros(Vm.shape)
```

```python
# I = np.ones(Sim_t.shape)

# I1 = fn.gen_sin(init_time=0,
#                 freq=0.01,
#                 phase=-0,
#                 time_points=Sim_t)
# I2 = fn.gen_sin(init_time=0,
#                 freq=0.011,
#                 phase=-0,
#                 time_points=Sim_t)

# I = I1 + I2

I_stim = 100

# I = I_stim * np.ones(t.shape)  # pA

# Chirp stimulation
out, freq = fn.gen_chirp(Chirp_init_freq=.00000001, 
                         Chirp_init_time=0, 
                         Chirp_end_freq=.01, 
                         Chirp_end_time=Sim_t[-1], 
                         time_points=Sim_t)

I      = I_stim * out  # pA
```

```python
for i in range(I.shape[0]-1):
    
    # diode non-linearity
    if Vm[i] > V_half:
        diode_nl[i] = 1
    else:
        diode_nl[i] = 0
    
    # current through the membrane capacitanc    
    dVmdt = (I[i] - g_leak*(Vm[i]-0) - g_Na*diode_nl[i]*(Vm[i]-0)) / C
    Vm[i+1] = Vm[i] + dVmdt * Sim_dt
```

```python
plt.plot(Sim_t, I)
plt.show()
plt.plot(Sim_t, Vm)
plt.show()
plt.plot(Sim_t, diode_nl)
plt.show()
```

### Demodulator using Boltzmann non-linearity

```python
Vm = np.zeros(Sim_t.shape)
```

Parameetrs of the Boltzmann function

```python
V_half = 0
k = 1
Boltzman_nl = np.zeros(Vm.shape)
```

```python
for i in range(I.shape[0]-1):
    
    # diode non-linearity
    Boltzman_nl[i] = k / (1 + np.exp(V_half - Vm[i]))
    
    # current through the membrane capacitanc
    I_leak = g_leak*(Vm[i]-0)
    I_Na = g_Na*(Boltzman_nl[i]**3)*(Vm[i]-0)
    
    dVmdt = (I[i] - I_leak - I_Na) / C
    Vm[i+1] = Vm[i] + dVmdt * Sim_dt
```

```python
plt.plot(Sim_t, I)
plt.show()
plt.plot(Sim_t, Vm)
plt.show()
plt.plot(Sim_t, Boltzman_nl)
plt.show()
```

### Adding the inductor

```python
g_leak = 1
g_Na = .01
g_L = 1
L = .01
C = 100
V_leak = 0
V_Na = 0

tau_L = 1 / (g_L * L)
print(tau_L)
```

```python
Vm = np.zeros(Sim_t.shape)
I_L = np.zeros(Sim_t.shape)
```

```python
V_half = 0
k = 1
Boltzman_nl = np.zeros(Vm.shape)
```

```python
for i in range(I.shape[0]-1):
    
    # Boltzmann non-linearity
    Boltzman_nl[i] = k / (1 + np.exp(V_half - Vm[i]))
    
    # current through the leak branch
    I_leak = g_leak*(Vm[i]-V_leak)
    
    # current through the membrane capacitance
    I_Na = g_Na*(Boltzman_nl[i]**3)*(Vm[i]-V_Na)
    
    # current through the inductor
    I_L_ss = Vm[i] * g_L
    dILdt = (I_L_ss - I_L[i]) / tau_L
    I_L[i+1] = I_L[i] + dILdt * Sim_dt
    
    dVmdt = (I[i] - I_leak - I_Na - I_L[i]) / C
    Vm[i+1] = Vm[i] + dVmdt * Sim_dt
```

```python
plt.plot(Sim_t, I)
plt.show()
plt.plot(Sim_t, Vm)
plt.show()
plt.plot(Sim_t, I_L)
plt.show()
plt.plot(Sim_t, Boltzman_nl)
plt.show()
```
