```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt
import numpy as np
import functions as fn
```

```python
t = np.linspace(0, 10, 5001)
w = chirp(t, f0=.1, f1=10, t1=10, method='linear')
plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()
```

```python
s = fn.gen_sin(freq=1, time_points=t)
plt.plot(t, s)
plt.title("sine wave")
plt.xlabel('t (sec)')
plt.show()
```

```python
t = np.linspace(0, 20, 500001)
w = chirp(t, f0=1000, f1=1020, t1=t[-1], method='linear')
s = fn.gen_sin(freq=1000, time_points=t)

plt.figure(figsize=(20,5))
plt.plot(t, w+s)
plt.show()
```

```python
t = np.linspace(0, 20, 500001)
w, freq = fn.gen_chirp(Chirp_init_freq=1000, Chirp_end_freq=1020, Chirp_end_time=t[-1], time_points=t)
s = fn.gen_sin(freq=1000, time_points=t)

plt.figure(figsize=(20,5))
plt.plot(t, w+s)
plt.show()
```
